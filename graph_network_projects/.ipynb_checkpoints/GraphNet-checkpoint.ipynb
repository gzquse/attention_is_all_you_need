# show concept of gcn with numpy

```python
# define adjacent matrix,  
A = np.matrix([
    [0, 1, 0, 0],
    [0, 0, 1, 1],
    [0, 1, 0, 0],
    [1, 0, 1, 0]],
    dtype=float
)
```

```python
# define node attribute
X = np.matrix([[i, -i] for i in range(A.shape[0])], dtype=float)
(A * X).shape
```

```python
# prepare adjacent matrix with self connection  
I = np.matrix(np.eye(A.shape[0]))
A_hat = A + I
A_hat * X
```

```python
# define degree matrix of A hat
D_hat = np.array(np.sum(A_hat, axis=0))[0]
D_hat = np.matrix(np.diag(D_hat))
# cinyte D^-1/2 A D^-1/2, because symmetric matric, permutable
D_hat** -1 * A_hat
```

```python
# the graph propagation process, from node to message, W will be a linear
# transformation to change dimension of features 
W = np.matrix([[1], [-1]])
D_hat**-1 * A_hat * X * W
```

```python
# the last step, node nonlinear transformation, transform original feature 
# to new node features 
# think why we get this final results? 

def relu(x):
    return (abs(x) + x) / 2
relu(D_hat**-1 * A_hat * X * W)
```

# Train gcn with pytorch

We try to classify two kinds of people by gcn. 

```python
# prepare graph
zkc = nx.karate_club_graph()
attr = {n: {"label": int(zkc.nodes[n]['club'] == 'Officer')} for n in zkc.nodes()}
nx.set_node_attributes(zkc, attr)
labels = [zkc.nodes[i]['label'] for i in range(34)]
order = sorted(list(zkc.nodes()))
```

```python
# prepare train data, get full graph neural network with pytorch
class GraphNet():
    def __init__(self, G, n_in, n_hid):
        # get labes for training
        self.labels = [G.nodes[i]['label'] for i in range(len(G.nodes))]
        order = sorted(list(G.nodes()))
        # get adjacent matrix 
        self.A = torch.from_numpy(nx.to_numpy_array(G, nodelist=order)).double()
        # get the degree matrix 
        self.I = torch.eye(G.number_of_nodes()).double()
        # combine A and I 
        self.A_hat = self.A + self.I
        self.D_hat = torch.diag(torch.sum(self.A_hat, dim=0))
        # feature transform matrix
        self.W = nn.Parameter(torch.randn([n_in, n_hid]).double())
        self.W_hid = nn.Parameter(torch.randn([n_hid, n_hid]).double())
        # feature transform matrix
        self.W_out = nn.Parameter(torch.randn([n_hid, 1]).double())
    # define gcn forward phase     
    def gcn_layer(self, A_hat, D_hat, X, W):
        return torch.relu(torch.inverse(D_hat).mm(A_hat).mm(X).mm(W)) 
    
    def forward(self, X, iters, hidden = True):
        if hidden == True:
            for i in range(iters):
                if i == 0:
                    X = self.gcn_layer(self.A_hat, self.D_hat, X, self.W)
                else:
                    X = self.gcn_layer(self.A_hat, self.D_hat, X, self.W_hid)
        output = self.gcn_layer(self.A_hat, self.D_hat, X, self.W_out)
        return output
    
        
Net = GraphNet(zkc, len(zkc.nodes), 1)
output = Net.forward(Net.I, 1)
output.shape
```

```python
# initial state, visuallize the output and labels, the initilize output
# show no information about correct label
Att = np.array([(a.data.numpy(), m) for a, m in zip(output, Net.labels)])
pos = nx.spring_layout(zkc, iterations=500)
nx.draw(zkc, pos, node_color = Att[:, 0], with_labels = True)
plt.figure()
nx.draw(zkc, pos, node_color = Att[:, 1], with_labels = True)
```

```python
# write the training process    
Net = GraphNet(zkc, len(zkc.nodes), 1)
Loss = torch.nn.BCELoss()
optim = torch.optim.Adam([Net.W, Net.W_hid, Net.W_out], lr = 1e-2)
target = torch.FloatTensor(labels).reshape(-1).double()
# we can choose different input attribute, but should be fixed before loops of training
# also it seems that it cannot be identical node input
arr = np.arange(34)
np.random.shuffle(arr)
X = Net.I[arr]
# X = torch.randn([zkc.number_of_nodes(), zkc.number_of_nodes()])
# X = torch.ones([zkc.number_of_nodes(), zkc.number_of_nodes()])
for i in range(5000):

    output = Net.forward(X.double(), 1)
    predict = output.reshape(-1)
    loss = Loss(torch.sigmoid(predict), target)
    loss.backward()
    if i%500 == 0:
        print(loss)
    optim.step()
    optim.zero_grad()
```

# Another view, deep walk

```python
# use the deepwalk to generate different sequence of nodes 
def get_randomwalk(G, node, path_length): 
    # initialize node from input
    random_walk = [str(node)]
    # try to walk in graph without going back to origin
    for i in range(path_length-1): 
        # choose a node from neighbors, but never repeat
        temp = list(G.neighbors(node)) 
        temp = list(set(temp) - set(random_walk)) 
        if len(temp) == 0: 
            break 
        # pick up from uniform distribution 
        random_node = random.choice(temp) 
        random_walk.append(str(random_node))
        # update current state
        node = random_node 
    return random_walk
```

```python
# generate 1000 such words 
Words = []
A = []
for i in range(1000):
    random_walk = get_randomwalk(zkc, np.random.randint(0, 34), 10)
    Words.append(random_walk)
    A.extend(random_walk)
print(str(random_walk))
np.unique(A)
```

```python
import gensim 
from gensim.models import Word2Vec
```

```python
# try to embed all the possible word sequence, in order
model = Word2Vec(window = 8, sg = 1, hs = 0,
                 negative = 10, # for negative sampling
                 alpha=0.03, min_alpha=0.0007,
                 seed = 14)

model.build_vocab(Words, progress_per=2)
```

```python
model.train(Words, total_examples = model.corpus_count, epochs=20, report_delay=1)
```

```python
model.similar_by_word('3')
```

```python
# visualize the " word embedding" for different words
from sklearn.decomposition import PCA
def plot_nodes(word_list):
    X = model[word_list]
    
    # reduce dimensions to 2
    pca = PCA(n_components=2)
    result = pca.fit_transform(X)
    
    
    plt.figure(figsize=(12,9))
    # create a scatter plot of the projection


    for i, word in enumerate(word_list):
        plt.annotate(word, xy=(result[i, 0], result[i, 1]))
        if zkc.nodes[i]['label'] == 0:
            plt.scatter(result[i, 0], result[i, 1], color = 'r')
        else:
            plt.scatter(result[i, 0], result[i, 1], color = 'k')
    plt.show()
```

```python
# now we can see different labels belongs to different embedded regions
plot_nodes([str(a) for a in np.arange(34)])
```

# Ａ quick way for graph nets
There are mature packages for graph network, we can use torch.geometric for instance  


```python
import os.path as osp
import torch
import torch.nn.functional as F
import torch_geometric
from torch_geometric.data import Data
import torch_geometric.transforms as T
from torch_geometric.nn import GCNConv
```

```python
# transform network data to gemoetry form 
send = [z[0] for z in zkc.edges]
rec = [z[1] for z in zkc.edges]
edge_index = torch.tensor([[send, rec]])
x = torch.eye(zkc.number_of_nodes())
data = Data(x=x, edge_index=edge_index[0])
data.train_mask = torch.tensor(len(x) * [True])
data.y = torch.tensor(labels)
data.x
```

```python
# define gcn as simple as possible
class Net(torch.nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        # int channels out channels of gcn-network is just the input dim and output dim of node dimension  
        self.conv1 = GCNConv(34, 2)
        # On the Pubmed dataset, use heads=8 in conv2.
        self.conv2 = GCNConv(2, 1)

    def forward(self, x):
        x = F.dropout(x, p=0.6, training=self.training)
        x = F.elu(self.conv1(x, data.edge_index))
        x = F.dropout(x, p=0.6, training=self.training)
        x = self.conv2(x, data.edge_index)
        x = torch.sigmoid(x)
        return x
```

```python
I = torch.eye(zkc.number_of_nodes())
loss = torch.nn.BCELoss()
print (net(I)[data.train_mask].view(-1), data.y[data.train_mask].view(-1))
loss(net(I)[data.train_mask].view(-1), data.y[data.train_mask].view(-1).float())
```

```python
net = Net()
optimizer = torch.optim.Adam(net.parameters(), lr=0.005, weight_decay=5e-4)
def train():
    net.train()
    optimizer.zero_grad()
    Loss = torch.nn.BCELoss()
    loss = Loss(net(I)[data.train_mask].view(-1), data.y[data.train_mask].view(-1).float())
    loss.backward()
    optimizer.step()
    return loss
    

def test():
    net.eval()
    p = net(I)
    pred = (p>0.5)
#     print (pred, data.y)
    acc = pred.view(-1).eq(data.y).sum().item()/34
    return acc

for epoch in range(1, 1601):
    loss = train()
    log = 'Epoch: {:03d}, Train: {:.4f}, Val: {:.4f}, Test: {:.4f}'
    if epoch%80 == 0:
        print (loss)
        print ('test loss', test())
#     print(log.format(epoch, *test()))
```
